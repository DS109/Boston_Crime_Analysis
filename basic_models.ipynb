```python
crime_df = crime_df.dropna(subset=['wards'])
streetlight_df = streetlight_df.dropna(subset=['wards'])
```

```python
wards_li = []
for i in crime_df['wards'].values: 
    wards_li.append([int(i)])
crime_df['wards'] = pd.DataFrame(np.array(wards_li))

wards_li = []
for i in streetlight_df['wards'].values: 
    wards_li.append([int(i)])
streetlight_df['wards'] = pd.DataFrame(np.array(wards_li))
```

```python
'''
When the 'OCCURRED_ON_DATE' is between Nov. 3 and March 8, it will be belong to 'winter time'.
And only when the time is in the range of lighting-up, it will be counted into vaild data.
(for streetlight-related model).
'''
winter_time = []
light_up = []
for i,j in zip(crime_df['OCCURRED_ON_DATE'],crime_df['HOUR']):
    li = i.split('/')
    day = int(li[0]) * 100 + int(li[1])
    if 308 >= day or day >= 1103:
        winter_time.append([1])
        if 0 <= int(j) < 6 or 16 <= int(j) <= 24:
            light_up.append([1])
        else:
            light_up.append([0])
    else:
        winter_time.append([0])
        if 0 <= int(j) < 6 or 17 <= int(j) <= 24:
            light_up.append([1])
        else:
            light_up.append([0])
            
crime_df['WINTER_TIME'] = pd.DataFrame(np.array(winter_time))
crime_df['LIGHT_UP'] = pd.DataFrame(np.array(light_up))
crime_df
```

```python
def t_test(col_1,col_2):
    
    mean_1 = col_1.mean()
    std_1 = col_1.std()
    n_1 = len(col_1)
    
    mean_2 = col_2.mean()
    std_2 = col_2.std()
    n_2 = len(col_2)
    
    return np.abs(mean_1-mean_2) / (np.sqrt(std_1 **2 / n_1 + std_2 **2 / n_2))

def normalize(col):
    max_col = col.max()
    min_col = col.min()
    delta = max_col - min_col
    print(delta)
    col = (col - min_col)/(delta)
    return col
```

```python
type_li = np.zeros((len(crime_df['OFFENSE_CODE_GROUP']),1), dtype=int)
crime_df_co = crime_df.copy()
for i in range(6):
    for j in crime_df_co['class_'+str(i)]:
        if j == 1:
            type_li[i] = i
crime_df_co['TYPE'] = pd.DataFrame(type_li)
crime_df_co['TYPE'].unique()
```

```python
minute_li = []
for i in crime_df['OCCURRED_ON_DATE'].values:    
    minute_li.append([int(i[-2:])])
crime_df_co['MINUTE'] = pd.DataFrame(np.array(minute_li))

day_li = []
for i in crime_df['DAY_OF_WEEK'].values: 
    if i == 'Sunday':
        day_li.append([0])
    if i == 'Monday':
        day_li.append([1])
    if i == 'Tuesday':
        day_li.append([2])
    if i == 'Wednesday':
        day_li.append([3])
    if i == 'Thursday':
        day_li.append([4])
    if i == 'Friday':
        day_li.append([5])
    if i == 'Saturday':
        day_li.append([6]) 
crime_df_co['DAY_OF_WEEK'] = pd.DataFrame(np.array(day_li))
crime_df_co.drop(['OFFENSE_CODE_GROUP','OCCURRED_ON_DATE','Lat','Long','WINTER_TIME'],inplace=True)
crime_df_co
```

```python
t_test_params = ['LIGHT_UP','MONTH','YEAR','DAY_OF_WEEK','HOUR']
for i in t_test_params:
    crime_df_co[i] = normalize(crime_df_co[i])
    type_with_light = t_test(crime_df_co['TYPE'],crime_df_co[i])
    print(i,type_with_light)
    break

crime_df_co
```

```python
crime_df_co = crime_df_co[crime_df_co['LIGHT_UP']==1]
crime_wards_df = pd.DataFrame(crime_df_co['wards'].value_counts())
crime_wards_df.columns = ['NUM_OF_CRIME']
crime_wards_df = crime_wards_df.sort_index()
crime_wards_df
```

```python
streetlight_df_co = streetlight_df.copy()
streetlight_df_co.drop(['Lat','Long'],axis=1,inplace=True)
streetlight_wards_df = pd.DataFrame(streetlight_df_co['wards'].value_counts())
streetlight_wards_df.columns = ['NUM_OF_LIGHTS']
streetlight_wards_df = streetlight_wards_df.sort_index()
streetlight_wards_df['SHAPESTArea'] = wards_df['SHAPESTArea'] / (1e+7)
streetlight_wards_df['DENSITY'] = streetlight_wards_df.apply(lambda x: x['NUM_OF_LIGHTS'] / x['SHAPESTArea'], axis=1)
streetlight_wards_df.drop(['SHAPESTArea'],axis=1,inplace=True)
streetlight_wards_df
```

```python
X_train, X_test, y_train, y_test = train_test_split(streetlight_wards_df['DENSITY'].values.reshape(-1,1),crime_wards_df['NUM_OF_CRIME'].values.reshape(-1,1),test_size=0.2, random_state=42)

streetlight_to_crime_linear = LinearRegression().fit(X_train,y_train)
print('linear model score in train data: ',streetlight_to_crime_linear.score(X_train,y_train))
print('linear model score in test data: ',streetlight_to_crime_linear.score(X_test, y_test))

plt.figure(figsize=(15,8))
plt.scatter(X_train, y_train,label='train data')
plt.scatter(X_test, y_test,label='test data')
plt.plot(X_train,streetlight_to_crime_linear.predict(X_train),label='linear model')
plt.xlabel('streetlight number')
plt.ylabel('crime number')
plt.legend()
plt.show()
```
